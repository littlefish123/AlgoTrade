```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, LSTM
from xgboost import XGBRegressor
from sklearn.metrics import mean_squared_error, r2_score

# Load the data
df = pd.read_csv("https://datatraining.site/data/GLE-PA-data.csv")

# Select relevant features
features = ["Open", "High", "Low", "Close", "Volume"]
df = df[features]

# Preprocess the data
scaler = MinMaxScaler(feature_range=(0, 1))
df[features] = scaler.fit_transform(df[features])

# Create lagged features
window_size = 10
for i in range(1, window_size + 1):
    # df[f"Open_lag_{i}"] = df["Open"].shift(i).fillna(method="bfill")
    # df[f"High_lag_{i}"] = df["High"].shift(i).fillna(method="bfill")
    # df[f"Low_lag_{i}"] = df["Low"].shift(i).fillna(method="bfill")
    # df[f"Close_lag_{i}"] = df["Close"].shift(i).fillna(method="bfill")
    # df[f"Volume_lag_{i}"] = df["Volume"].shift(i).fillna(method="bfill")

    df[f"Open_lag_{i}"] = df["Open"].shift(i).bfill()
    df[f"High_lag_{i}"] = df["High"].shift(i).bfill()
    df[f"Low_lag_{i}"] = df["Low"].shift(i).bfill()
    df[f"Close_lag_{i}"] = df["Close"].shift(i).bfill()
    df[f"Volume_lag_{i}"] = df["Volume"].shift(i).bfill()


# Split the data into training and testing sets
train_data, test_data = train_test_split(df, test_size=0.2, shuffle=False)

# Prepare data for CNN
X_train_cnn = np.array(train_data[features].values)
y_train_cnn = np.array(train_data["Close"].values)
X_test_cnn = np.array(test_data[features].values)
y_test_cnn = np.array(test_data["Close"].values)

# Reshape data for CNN
X_train_cnn = X_train_cnn.reshape((X_train_cnn.shape[0], X_train_cnn.shape[1], 1))
X_test_cnn = X_test_cnn.reshape((X_test_cnn.shape[0], X_test_cnn.shape[1], 1))

# Build CNN model
cnn_model = Sequential()
cnn_model.add(Conv1D(filters=32, kernel_size=3, activation='relu', input_shape=(X_train_cnn.shape[1], 1)))
cnn_model.add(MaxPooling1D(pool_size=2))
cnn_model.add(Flatten())
cnn_model.add(Dense(1, activation='linear'))
cnn_model.compile(optimizer='adam', loss='mse')

# Train CNN model
cnn_model.fit(X_train_cnn, y_train_cnn, epochs=50, batch_size=32, verbose=1)

# Predict using CNN
y_pred_cnn = cnn_model.predict(X_test_cnn)

# Prepare data for XGBoost
X_train_xgb = train_data.drop("Close", axis=1)
y_train_xgb = train_data["Close"]
X_test_xgb = test_data.drop("Close", axis=1)
y_test_xgb = test_data["Close"]

# Build XGBoost model
xgb_model = XGBRegressor(n_estimators=100, learning_rate=0.1, max_depth=5)

# Train XGBoost model
xgb_model.fit(X_train_xgb, y_train_xgb)

# Predict using XGBoost
y_pred_xgb = xgb_model.predict(X_test_xgb)

# Evaluate models
mse_cnn = mean_squared_error(y_test_cnn, y_pred_cnn)
r2_cnn = r2_score(y_test_cnn, y_pred_cnn)
mse_xgb = mean_squared_error(y_test_xgb, y_pred_xgb)
r2_xgb = r2_score(y_test_xgb, y_pred_xgb)

print(f"CNN - MSE: {mse_cnn:.4f}, R-squared: {r2_cnn:.4f}")
print(f"XGBoost - MSE: {mse_xgb:.4f}, R-squared: {r2_xgb:.4f}")

## ... (rest of the code)

# ... (rest of the code)

# Scale only the "Close" column
close_scaler = MinMaxScaler(feature_range=(0, 1))
train_data["Close"] = close_scaler.fit_transform(train_data["Close"].values.reshape(-1, 1))
test_data["Close"] = close_scaler.transform(test_data["Close"].values.reshape(-1, 1))

# ... (rest of the code)

# Predict using CNN
y_pred_cnn = cnn_model.predict(X_test_cnn)

# Inverse transform predictions
y_pred_cnn = close_scaler.inverse_transform(y_pred_cnn.reshape(-1, 1))

# ... (rest of the code)
# ... (rest of the code)
# Plot results
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.plot(y_test_cnn, label="Actual")
plt.plot(y_pred_cnn, label="CNN Prediction")
plt.legend()
plt.title("CNN Prediction vs Actual")
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(y_test_xgb, label="Actual")
plt.plot(y_pred_xgb, label="XGBoost Prediction")
plt.legend()
plt.title("XGBoost Prediction vs Actual")
plt.show()
```
